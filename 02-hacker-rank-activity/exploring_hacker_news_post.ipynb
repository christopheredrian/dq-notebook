# Exploring Hacker News Posts

Requirements
- Do Ask HN or Show HN receive more comments on average?
- Do posts created at a certain time receive more comments on average?

Input
- `hacker_news.csv`

```python
import csv as c
import datetime as dt
import pprint
pp = pprint.PrettyPrinter(indent=4)

ask_posts = []
show_posts = []
other_posts = []

total_ask_comments = 0
total_show_comments = 0


with open('hacker_news.csv') as file:
    
    reader = c.reader(file)
    hn = list(reader) # convert to list 
    
    # returns and remove first element from the list 
    headers = hn.pop(0) 
    
    for index, row in enumerate(hn, start=0):
        
        # columns
        title = row[1]
        title_lower = title.lower()
        
        num_comments =  int(row[4])
        
        if title_lower.startswith('ask hn'):
            ask_posts.append(row)
            total_ask_comments += num_comments 
        elif title_lower.startswith('show hn'): 
            show_posts.append(row)
            total_show_comments += num_comments 
        else:
            other_posts.append(row)

avg_ask_comments = total_ask_comments / len(ask_posts)
avg_show_comments = total_show_comments / len(show_posts)
            
print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
print(avg_ask_comments)
print(avg_show_comments)
print('done')


# Calculate the number of comments
result_list = []

for post in ask_posts:
    created_at = post[6]
    number_of_comments = int(post[4])
    result_list.append([created_at, number_of_comments])
    
count_by_hour = {}
comments_by_hour = {}

pp.pprint(result_list[0:5])
for result in result_list:
    created_at_dt = dt.datetime.strptime(result[0], "%m/%d/%Y %H:%M")
    number_of_comments = result[1]
    hour = created_at_dt.strftime("%H")
    
    if hour not in count_by_hour:
        count_by_hour[hour] = 1
        comments_by_hour[hour] = number_of_comments
    else:
        count_by_hour[hour] += 1
        comments_by_hour[hour] += number_of_comments
    
# print("Number of posts per hour")
# pp.pprint(count_by_hour)
# print("Comments per hour")
# pp.pprint(comments_by_hour)
avg_by_hour = []
## Calculate the average number of comments for posts 
## created during each hour of the day.
# avg_comments_per_hour = [] # [<<hour>>, <<average number of comments per hour>>]

for hour in count_by_hour:
    avg_by_hour.append([hour, comments_by_hour[hour] /  count_by_hour[hour]])

print
## Average Comments per hour
# print("Average Comments per Hour")
# pp.pprint(avg_by_hour)
swap_avg_by_hour = []

for list_item in avg_by_hour:
    swap_avg_by_hour.append([list_item[1], list_item[0]])

sorted_swap = sorted(swap_avg_by_hour, reverse=True)

# pp.pprint(sorted_swap)
print("Top 5 Hours for Ask Posts Comments")

for top_five_contestant in sorted_swap[:5]:
    formatted_date = dt.datetime.strptime(top_five_contestant[1], "%H").strftime("%H:00")
    print("{}: {:.2f} average comments per post".format(formatted_date, top_five_contestant[0]))
```
